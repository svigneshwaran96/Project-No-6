```python
plt.bar(x,y,color= 'Red', width=0.3,edgecolor = 'Blue',align = 'edge')
plt.show()
```

```python
plt.bar(x,y,color= 'Red', width=0.3,edgecolor = 'Blue',align = 'edge')
plt.title('My first bar chart')
plt.show()
```

```python
plt.bar(x,y,color= 'Red', width=0.3,edgecolor = 'Blue',align = 'edge')
plt.title('My first bar chart')
plt.xlabel('Fruits')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
plt.pie([23,45,11,3,98],labels=['Apple','Banana','Jack','Mango','Orange'])
plt.show()
```

```python
plt.barh(x,y,color= 'Red')
plt.title('My first bar chart')
plt.xlabel('Fruits')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.scatterplot(data=df,x ='tip', y ='total_bill' )
```

```python
sns.scatterplot(data=df,x ='tip', y ='total_bill' ,hue = 'sex')
```

```python
sns.scatterplot(data=df,x ='tip', y ='total_bill' ,hue = 'sex', size = 'size')
```

```python
sns.scatterplot(data=df,x ='tip', y ='total_bill' ,hue = 'sex', size = 'size',style = 'smoker')
```

```python
sns.relplot(data=df,x ='tip', y ='total_bill' ,hue = 'sex', size = 'size',style = 'smoker',row = 'day')
```

```python
sns.relplot(data=df,x ='tip', y ='total_bill' ,hue = 'sex', size = 'size',style = 'smoker',row = 'day',col = 'time')
```

```python
sns.boxplot(df['total_bill'])
```

```python
sns.pairplot(df)
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.violinplot(df['total_bill'])
```

```python
sns.jointplot(data=df,x='tip',y='total_bill')
```

```python
sns.swarmplot(df['total_bill'])
```
